```python
from transformers import AutoModelForMaskedLM, AutoModelForQuestionAnswering
from transformers import pipeline, AutoTokenizer, FillMaskPipeline, QuestionAnsweringPipeline
from pprint import pprint
```

```python
#tokenizer_hf = AutoTokenizer.from_pretrained('PlanTL-GOB-ES/roberta-base-bne')
#model = AutoModelForMaskedLM.from_pretrained('PlanTL-GOB-ES/roberta-base-bne')
tokenizer_hf = AutoTokenizer.from_pretrained('PlanTL-GOB-ES/roberta-base-bne-sqac')
model = AutoModelForQuestionAnswering.from_pretrained('PlanTL-GOB-ES/roberta-base-bne-sqac')
model.eval()
```

```python
#pipeline = FillMaskPipeline(model, tokenizer_hf)
# pipeline = QuestionAnsweringPipeline(model, tokenizer_hf)
pipeline = pipeline(task="question-answering", model=model, tokenizer=tokenizer_hf)
```
